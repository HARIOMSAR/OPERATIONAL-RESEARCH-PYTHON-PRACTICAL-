1.Write a program to plot a bar chart in python to display the result of a school for five consecutive years.

```python
import matplotlib.pyplot as plt
years=["2018","2019","2020","2021","2022"]
results=[80,85,89,95,95]
plt.bar(years,results)
plt.title("School result for five consecutive years")
plt.xlabel("Years")
plt.ylabel("Results")
plt.show()
```

2.Write a program in python to plot a graph for the function y=x^2

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return x**2
x=np.linspace(-5,5,100)
y=f(x)
plt.plot(x,y)
plt.title("y=x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

3.Write a program in python to plot a pie chart on consumption of water in daily life. 

```python
import matplotlib.pyplot as plt
labels=["Drinking","Cooking","Bathing","Cleaning"]
sizes=[2,1,3,4]
fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,startangle=90)
ax1.axis('equal')
plt.title("Water consumption in daily life")
plt.show
```
